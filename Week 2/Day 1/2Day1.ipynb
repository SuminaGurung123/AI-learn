#Measure of central tendency


we can also do by statistic module

```python
import pandas as pd
a=pd.Series([1,2,3,4,4,4])

mean=a.mean()
print(mean)

median=a.median()
print(median)

mode=a.mode()[0]
print(mode)
```

```python
  data=pd.Series([1,2,3,4,4,4])
  range=max(data)-min(data)
  print(range)
```

```python
 data=pd.Series([1,2,3,4,4,4])
 variance=data.var()
 std=data.std()
 print(variance)
 print(std)
```

#Data visualizing

```python
import matplotlib.pyplot as plt
import pandas as pd

data=pd.DataFrame({'x':[-10,1,2,3,4,5,2,4,6,8,10,100]})
plt.boxplot(data['x'])
plt.title("box polt")
plt.xlabel("X")
plt.ylabel("value")
plt.show()
```
